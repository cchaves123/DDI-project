```python
N = 100
p = 1000
alpha = 1000
alpha2 = -1000
# beta_star =  np.concatenate([alpha * np.ones(p), alpha2 * np.ones(p)])
##
beta_star = np.array(np.arange(1, 1000, 1))
X = np.random.normal(loc=1, scale=1, size= (N, len(beta_star)))
Y = X.dot(beta_star)
```

```python
import matplotlib.pyplot as plt
plt.hist(Y)
```

```python
import sklearn as sk
```

```python
from sklearn.linear_model import LassoCV, ElasticNet
```

```python
reg = LassoCV(cv=5, random_state=0, fit_intercept=False).fit(X, Y)
```

```python
reg.score(X, Y)
```

```python
plt.scatter(beta_star, reg.coef_)
```

```python
plt.hist(np.abs(beta_star- reg.coef_))
```

```python
plt.hist(reg.coef_)
```

```python
reg2 = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=0, fit_intercept=False).fit(X, Y)
```

```python
plt.hist(reg2.coef_)
```

```python
plt.scatter(Y,reg2.predict(X))
```

```python
plt.hist(Y- reg2.predict(X))
```
